The process dealing with each year's 311 data (2014 to 2019)includes these steps: 1. import the csv file and convert it into a datafram. The name is "flooding + the year". 2. Format the header, select rows about flooding (key words "Catch Basin - Blocked / Flooding") and parse the date&time column. 3. Regroup the dataframe by counting how many complaints were received each day. Generate another dataframe with name pattern "counts + two-digit year". 4. Visualize the new dataframe and compare it with the annual percipitation data. 

```python
import pandas as pd
```

```python
import matplotlib.dates as mdates
```

```python
%matplotlib inline
```

### 2014 data starts here

```python
flooding2014 = pd.read_csv('SR2014.csv')
```

```python
flooding2014.columns = flooding2014.columns.str.strip().str.lower().str.replace(' ', '_').str.replace('(', '').str.replace(')', '')
```

```python
flooding2014 = flooding2014[flooding2014.service_request_type =='Catch Basin - Blocked / Flooding']
```

```python
flooding2014['dates'] = pd.to_datetime(flooding2014['creation_date']).dt.date
```

```python
flooding2014['time'] = pd.to_datetime(flooding2014['creation_date']).dt.time
```

```python
flooding2014['year'] = pd.to_datetime(flooding2014['creation_date']).dt.year
```

```python
counts14 = flooding2014.dates.value_counts().reset_index()
```

```python
counts14 = counts14.rename(index=str, columns={"index": "date", "dates": "counts"})
```

```python
counts14 = counts14.sort_values('date')
```

```python
chart14 = counts14.plot.line(x='date', y='counts', figsize=(15,7)).xaxis.set_major_formatter(mdates.DateFormatter('%b %d'))
```

Import the historical weather data which is available from the Government of Canada's website (http://climate.weather.gc.ca/historical_data/search_historic_data_e.html).

```python
percip14 = pd.read_csv('percip2014.csv')
```

```python
percip14['Date/Time'] = pd.to_datetime(percip14['Date/Time']).dt.date
```

```python
percip14.plot.line(x='Date/Time', y='Total_Precip', figsize=(15,7)).xaxis.set_major_formatter(mdates.DateFormatter('%b %d'))
```

## 2018 data starts here

```python
flooding2018 = pd.read_csv('SR2018.csv')
```

```python
flooding2018.columns = flooding2018.columns.str.strip().str.lower().str.replace(' ', '_').str.replace('(', '').str.replace(')', '')
```

```python
flooding2018 = flooding2018[flooding2018.service_request_type =='Catch Basin - Blocked / Flooding']
```

```python
flooding2018['dates'] = pd.to_datetime(flooding2018['creation_date']).dt.date
```

```python
flooding2018['time'] = pd.to_datetime(flooding2018['creation_date']).dt.time
```

```python
flooding2018['year'] = pd.to_datetime(flooding2018['creation_date']).dt.year
```

```python
counts18 = flooding2018.dates.value_counts().reset_index()
```

```python
counts18 = counts18.rename(index=str, columns={"index": "date", "dates": "counts"})
```

```python
counts18 = counts18.sort_values('date')
```

```python
counts18['date'] = pd.to_datetime(counts18['date'])
```

```python
idx = pd.date_range(start = '2018-01-01', end = '2018-12-31', freq = 'D')
```

```python
counts18 = counts18.set_index('date').reindex(idx).ffill().reset_index().rename(columns={'index': 'date'})
```

```python
percip18 = pd.read_csv('percip2018.csv')
```

```python
percip18['Date/Time'] = pd.to_datetime(percip18['Date/Time'])
```

```python
percip18 = percip18.rename(columns={'Date/Time':'date'})
```

```python
merged18 = pd.merge(percip18, counts18, on='date')
```

```python
ax = merged18.plot('date', 'counts')
merged18.plot('date','Total_Precip',secondary_y=True, ax=ax, figsize=(15,7))
```
